## Manipulation et prétraitement de données

### Premières manipulations d’un jeu de données

```python
# Importation des bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msn
```

1. Importez ce jeu de données au format csv

```python
df_age = pd.read_csv("age_gender.csv", encoding='utf-8', sep = ';')
df_age.head()
```

```python
print("Info", df_age.info())
print("_______________________________")
print("Describe", df_age.describe())
print("_______________________________")
print("Shape", df_age.shape)
```

2. Utilisez une fonction pour couper l’âge des individus par morceaux de cinq années (15 à 20, 20 à 25, etc.), puis affichez la distribution via matplotlib ou ggplot2. Utilisez la fonction cut (sur R ou sur pandas de Python)

```python
# Création des groupes (intervalles) de 5 ans chacun, commençant à 0 
#et se terminant à 100
intervalles = range(0, 105, 5)

# Regroupement des âges dans des intervalles à l'aide de la fonction pd.cut()
df_age['age_group'] = pd.cut(df_age['age'], intervalles)

# Le résultat
df_age
```

```python
sns.countplot(data = df_age, y = 'age_group')
plt.xlabel('Count')
plt.ylabel('Age')
plt.title('Distribution des âges dans des intervalles')
plt.show()
```

3.  Représentez la distribution des deux variables, l’âge d’une part, et le genre d’autre part, sur un même graphique, sous la forme de deux subplots


### Option 1
Conversion de la colonne **age_group** en données **int** et utiliser cette colonne avec la colonne **genre**.

```python
# Compter le nombre d'observations pour age_group
age_counts_0 = df_age[df_age["genre"] == 0]["age_group"].value_counts()
age_counts_1 = df_age[df_age["genre"] == 1]["age_group"].value_counts()

# Conversion des données catégorielles en valeurs numériques et obtention
#des catégories (avec l'aide de Chat GPT 3).
x0 = age_counts_0.index.codes
x1 = age_counts_1.index.codes
labels = age_counts_0.index.categories

# Création du barplot 
plt.bar(x0, age_counts_0.values, alpha=0.5)
plt.bar(x1, age_counts_1.values, alpha=0.5)

# Création des labels sur l'axe des abscisses
plt.xticks(range(len(labels)), labels, rotation = 90)

# Création d'une légende
plt.legend(["Homme", "Femme"])
plt.title("Distribution de l'âge par genre")

# Le résultat
plt.show()
```

```python
sns.countplot(x="age_group", data=df_age,hue="genre")
plt.xticks(rotation = 90)
plt.legend(["Homme", "Femme"])
plt.show()
```

### Option 2
Utilisation de la colonne **genre** et de la colonne **âge** et création de deux histogrammes.

```python
sns.histplot(df_age[df_age["genre"]==0]["age"], bins=15, color='#32CD32')
sns.histplot(df_age[df_age["genre"]==1]["age"], bins=15, color='#FF69B4')
plt.legend(["Homme", "Femme"])
plt.show()
```

### Option 3
En utilisant la colonne d'**âge** et en utilisant le paramètre **hue**, créer un graphique avec la distribution des âges et du genre

```python
# On remplace les 1 et 0 de genre par Femme et Homme en utilisant la méthode map
df_age["genre"] = df_age["genre"].map({1: "Femme", 0: "Homme"})
df_age
```

```python
# Création de l'histogramme des âges répartis par genre
sns.histplot(data=df_age, x="age", bins=15, hue='genre', kde = True) 
sns.set_palette('muted')

# Création d'étiquettes et de titre
plt.xlabel('Age')
plt.ylabel('Count')
plt.title("Distribution de l'âge par genre")
plt.show()
```

### Option 4
Utilisation d'un FacetGrid et création de deux subplots, un pour chaque genre.

```python
sns.color_palette("pastel")
g = sns.FacetGrid(df_age, col="genre", height=5, aspect=1)
g.map(sns.histplot, "age", kde=True)
g.set_titles(col_template="{col_name}")
plt.show()
```

5.  Représentez dans un même graphique la distribution de l’âge des individus en faisant une courbe pour les hommes, une courbe pour les femmes (indice pour Python : hue - seaborn)


```python
sns.histplot(df_age[df_age["genre"]=='Homme']["age"], bins=15, color='#32CD32', 
             kde=True)
sns.histplot(df_age[df_age["genre"]=='Femme']["age"], bins=15, color='#FF69B4', 
             kde=True)  

plt.title("Distribution de l'âge selon le genre")
plt.legend(["Homme", "Femme"])
plt.show()
```

6. Faites le même graphique en jouant sur la transparence des barres (paramètre alpha) pour mieux visualiser l’intersection des deux distributions

```python
sns.histplot(df_age[df_age["genre"]=='Homme']["age"], bins=15, alpha=0.1,
             color='#32CD32', kde=True)
sns.histplot(df_age[df_age["genre"]=='Femme']["age"], bins=15, alpha=0.2,
             color='#FF69B4', kde=True)

plt.title("Distribution de l'âge selon le genre")
plt.legend(["Homme", "Femme"])
plt.show()
```

7. Faites un barplot représentant sur une barre l’âge moyen des hommes, et sur l’autre l’âge moyen des femmes. Vous ajouterez une barre d’erreur correspondant à l’écart-type

```python
sns.set(style="white")
# Opération de calcul de la moyenne et de l'écart type par genre
age_means = df_age.groupby('genre')['age'].mean()
age_std = df_age.groupby('genre')['age'].std()

colors = ['#32CD32', '#FF69B4']

# Création d'un graphique à barres
plt.bar(x=['Hommes', 'Femmes'], height=age_means, yerr=age_std,
        capsize=8, color = colors)

# Titre et libellés du graphique
plt.xlabel('Genre')
plt.ylabel('Age moyen')
plt.title('Age moyen par Genre')

# Résultat
plt.show()
```
